```python
import updated_pipeline as up
import pandas as pd
```

```python
#Read in dataset
data = pd.read_csv('projects_2012_2013.csv')

#Convert relevant columns to datetime
data['date_posted'] = pd.to_datetime(data.date_posted)
data['datefullyfunded'] = pd.to_datetime(data.datefullyfunded)
data.fillna(data.median(), inplace=True)

#Create target column based on chosen criteria
data['target'] = ((data['datefullyfunded'] - data['date_posted']).dt.days > 60) * 1

#Create categorical variables for data analysis
for col in ['school_metro', 'primary_focus_subject', 'primary_focus_area', 'secondary_focus_area',
            'secondary_focus_subject', 'secondary_focus_area', 'teacher_prefix',
            'school_state']:
    data[col] = data[col].astype('category').cat.codes

#Created selected dummy variables
for col in ['resource_type', 'poverty_level', 'grade_level']:
    data = up.create_binary_vars(data, col)

#Convert string columns to dummy variables
up.convert_tf(data, ['school_magnet', 'school_charter', 'eligible_double_your_impact_match'], 'f')

#Drop columns from feature set
feats = data.drop(['projectid', 'teacher_acctid', 'schoolid', 'school_ncesid', 'school_latitude',
                  'school_longitude', 'school_city', 'school_district', 'school_county', 'resource_type', 'poverty_level',
                  'grade_level', 'date_posted', 'datefullyfunded', 'target'], axis=1)

#Chosen parameters to run in classification model
parameters = { 
    'RF':{'n_estimators': [1, 10], 'max_depth': [1, 5, 10], 'max_features': ['sqrt'],'min_samples_split': [5, 10]},
    'LR': { 'penalty': ['l1'], 'C': [0.01, 0.1, 1]},
    'AB': { 'algorithm': ['SAMME'], 'n_estimators': [1, 10]},
    'GB': {'n_estimators': [1, 10], 'learning_rate' : [0.1],'subsample' : [0.5], 'max_depth': [1]},
    'DT': {'criterion': ['gini'], 'max_depth': [1, 5, 10],'min_samples_split': [10]},
    'SVM' :{'C' :[0.01],'kernel':['linear']},
    'KNN' :{'n_neighbors': [3, 5, 10],'weights': ['uniform'],'algorithm': ['auto']},
    'BG': {'n_estimators': [1,10,100], 'max_samples': [2, 4, 8, 16, 32]}
}

#Models we have chosen to run for DonorsChoose
models_to_run = ['RF', 'AB', 'LR', 'GB', 'DT', 'KNN', 'BG']

#Dataframe of evaluation metrics
results = up.classify(data, up.split_time(data, '2012-01-01', '2013-12-31', [6]), 'date_posted',
                   feats.columns.tolist(), 'target', parameters, models_to_run)
```

```python
#Evaluation metrics sorted by time period, threshold, and model type
results.groupby(['time_period', 'threshold','model_type']).mean()
```
