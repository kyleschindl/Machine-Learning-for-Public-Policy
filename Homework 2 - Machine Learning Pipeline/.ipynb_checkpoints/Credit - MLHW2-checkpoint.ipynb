## *Pre-Process Data:*

```python
'''
Here we simply replace missing values with the median of a column.
'''
pl.pre_process(credit)
```

## *Generate Features/Predictors:*

```python
'''
In this step we discretize 'MonthlyIncome' into quintiles in order
to deal with outliers. Additionally, we turn zipcode into dummy variables
'''
credit.MonthlyIncome = pl.discretize(credit.MonthlyIncome, 5)
credit = pl.create_binary_vars(credit, 'zipcode')
```

## *Build Classifier:*

```python
'''
Here we build a simple logistic regression model with a small
number of features.
'''
features = credit[['NumberOfTimes90DaysLate', 'NumberOfDependents',
                   'age', 'DebtRatio']]
target = credit.SeriousDlqin2yrs

y_test, y_pred = pl.classify_model(features, target)
```

## *Evaluate Classifier:*

```python
'''
The logistic regression model produced high accuracy and high
precision, however, this is largely due to the underlying dataset
being mostly false in the target variable. The recall is very low,
showing this model was not particularly useful.
'''
pl.evaluate_classifier(y_test, y_pred)
```
